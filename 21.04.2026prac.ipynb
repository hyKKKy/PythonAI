```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from ipywidgets import interact, widgets

df = pd.read_csv("job_salary_prediction_dataset.csv")
```

```python
display(df.info())
display(df.describe())
display(df.head())
```

```python
plot_type = widgets.Dropdown(
    options=["Line", "Bar", "Scatter"],
    value="Line",
    description="Plot:"
)

x_col = widgets.Dropdown(
    options=df.columns,
    description="X:"
)

y_col = widgets.Dropdown(
    options=df.select_dtypes(include="number").columns,
    description="Y:"
)

group_check = widgets.Checkbox(
    value=True,
    description="Do Group"
)

agg_func = widgets.Dropdown(
    options=["mean", "sum", "min", "max", "count"],
    value="mean",
    description="Function:"
)

sort_check = widgets.Checkbox(
    value=False,
    description="Sort"
)

save_button = widgets.Button(
    description="Save graph",
    button_style="success"
)

ui = widgets.HBox([
    plot_type,
    x_col,
    y_col,
    group_check,
    agg_func,
    sort_check,
    save_button
])
```

```python
def update(plot_type, x_col, y_col, group_check, agg_func, sort_check):

    plt.figure(figsize=(10, 6))

    match plot_type:
        case "Line":
            func = plt.plot
        case "Bar":
            func = plt.bar
        case "Scatter":
            func = plt.scatter
        case _:
            return

    if group_check:

        group = getattr(
            df.groupby(x_col)[y_col],
            agg_func
        )()

        if sort_check:
            group = group.sort_values()

        func(group.index, group.values)

    else:

        show_df = df

        if sort_check:
            show_df = show_df.sort_values(y_col)

        func(show_df[x_col], show_df[y_col])

    plt.title(f"{plot_type} plot of {y_col} by {x_col}")

    plt.xlabel(x_col)
    plt.ylabel(y_col)

    plt.xticks(rotation=90)
    plt.grid(True)

    plt.tight_layout()

    plt.show()

def save(button):

    plt.figure(figsize=(10, 6))

    if group_check.value:

        group = getattr(
            df.groupby(x_col.value)[y_col.value],
            agg_func.value
        )()

        if sort_check.value:
            group = group.sort_values()

        plt.bar(group.index, group.values)

    else:

        show_df = df

        if sort_check.value:
            show_df = show_df.sort_values(y_col.value)

        plt.bar(
            show_df[x_col.value],
            show_df[y_col.value]
        )

    plt.title("Saved Graph")

    plt.xticks(rotation=90)

    plt.tight_layout()

    plt.savefig("graph.png")

save_button.on_click(save)
```

```python
out = widgets.interactive_output(
    update,
    {
        "plot_type": plot_type,
        "x_col": x_col,
        "y_col": y_col,
        "group_check": group_check,
        "agg_func": agg_func,
        "sort_check": sort_check,
    }
)
```

```python
display(ui, out)
```
