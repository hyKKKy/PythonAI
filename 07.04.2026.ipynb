```python
import ipywidgets as widgets
from IPython.display import display
import pandas as pd

num1 = widgets.SelectionRangeSlider(options=range(100), index=(1, 9))
num2 = widgets.SelectionRangeSlider(options=range(100), index=(1, 9))
op = widgets.Dropdown(options=['*', '/', '+', '-'])

def calc(num1, num2, op):
    num1 = num1[0]
    num2 = num2[0]

    if op == '+':
        result = num1 + num2
    elif op == '-':
        result = num1 - num2
    elif op == '*':
        result = num1 * num2
    elif op == '/':
        result = num1 / num2 if num2 != 0 else None

    display(pd.DataFrame([[num1, num2, result]], columns=["num1","num2","Result"]))

res = widgets.interactive_output(calc, {
    'num1': num1,
    'num2': num2,
    'op': op
})

display(num1, num2, op, res)
```

```python
import ipywidgets as widgets
from ipywidgets import interact

banwords = ["bad", "nigga"]

def clean_text(text):
    words = text.split()

    filtered = [w for w in words if w.lower() not in banwords]

    return " ".join(filtered)

def show(text):
    result = clean_text(text)
    print("Processed text:")
    print(result)

interact(
    show,
    text=widgets.Text(
        description="Text:"
    )
)
```
